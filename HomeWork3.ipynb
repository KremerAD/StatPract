```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('populations.txt')
tdata = data.T
```

```python
#1 задание
plt.grid()
plt.plot(tdata[0],tdata[1],label = 'hare')
plt.plot(tdata[0],tdata[2],label = 'lynx')
plt.plot(tdata[0],tdata[3],label = 'carrot')
plt.xlabel('year')
plt.ylabel('population')
plt.legend(bbox_to_anchor = (1,1), loc = 'upper right',borderaxespad = 0.)
```

```python
#задание 2
plt.grid()
plt.plot(tdata[0], tdata[1]/(tdata[1] + tdata[2]) * 100, label = 'hare')
plt.plot(tdata[0], tdata[2]/(tdata[1] + tdata[2]) * 100, label = 'lynx')
plt.xlabel('year')
plt.ylabel('%')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0.)
```

```python
#задание 3
hare_mean = np.mean(tdata[1])
lynx_mean = np.mean(tdata[2])
carrot_mean = np.mean(tdata[3])
hare_std = np.std(tdata[1])
lynx_std = np.std(tdata[2])
carrot_std = np.std(tdata[3])
print('hare mean = ', hare_mean)
print('lynx mean = ', lynx_mean)
print('carrot mean = ', carrot_mean)
print('hare std = ', hare_std)
print('lynx std = ', lynx_std)
print('carrot std = ', carrot_std)
fig = plt.figure(figsize=(14,5))

plt.subplot(131)
plt.title(u'Hare Population', family='verdana')
plt.grid()
plt.hist(tdata[1], bins = 20, color = 'g', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(hare_mean, color='k', linestyle='dashed', linewidth=1)

plt.tight_layout()
plt.subplot(132)
plt.title(u'Lynx Population', family='verdana')
plt.grid()
plt.hist(tdata[2], bins = 20, color = 'red', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(lynx_mean, color='k', linestyle='dashed', linewidth=1)

plt.tight_layout()
plt.subplot(133)
plt.title(u'Carrot Population', family='verdana')
plt.grid()
plt.hist(tdata[3], bins = 20, color = 'orange', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(carrot_mean, color='k', linestyle='dashed', linewidth=1)
```

```python
#задание 4
plt.grid()
plt.plot(tdata[0], tdata[1], label = 'Hare')
plt.plot(tdata[0], tdata[2], label = 'Lynx')
plt.plot(tdata[0], tdata[3], label = 'Carrot')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.scatter(tdata[0][np.argmax(tdata[1])], np.max(tdata[1]))
plt.scatter(tdata[0][np.argmax(tdata[2])], np.max(tdata[2]))
plt.scatter(tdata[0][np.argmax(tdata[3])], np.max(tdata[3]))
```

```python
#задание 5
b = []
b = np.ndarray(b)
a = (tdata[1] > tdata[2]).astype(np.int)
b = (tdata[1] > tdata[2]).astype(np.str)
b = ' '.join(b)
b = b.replace("True", "H")
b = b.replace("False", "L")
b = b.split(' ')
species = 'Hares', 'Lynxes'
sizes = [sum(a), len(a) - sum(a)]
explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=species, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
