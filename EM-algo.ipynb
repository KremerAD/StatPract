```python
import numpy as np
from math import factorial
```

```python
y = np.array([7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 3, 1, 1, 2, 7, 3, 3, 2, 2, 2]) 
M=10
Eps = 1e-8
```

```python
x = np.random.rand()
pi_0 = np.array([x,1-x])
mu_0 = np.random.rand(2)
pi_1 = np.array([331,331])
mu_1 = np.array([331,331])
```

```python
def Func(s,i,j):
    n = factorial(s) // (factorial(i) * factorial(s-i))
    return n * j ** i * (1-j) ** (s-i)
```

```python
iter = 0
while np.abs(pi_1-pi_0).sum() + np.abs(mu_1-mu_0).sum() > Eps:
    print('iteration = ' + str(iter), 'pi = ' + str(pi_0), 'mu = ' + str(mu_0), sep='\n', end='\n\n')
    pi_1 = pi_0
    mu_1 = mu_0
    iter += 1
    r = []
    for i in y:
        l = []
        for j in mu_0:
            l.append(Func(M,i,j))
        r.append(l)
    r = np.array(r)
    r *= pi_0[None,:]
    r /= np.sum(r, axis = 1)[:,None]
    pi_0 = np.average(r, axis = 0)
    mu_0 = np.sum(y[:, None] * r, axis = 0) / (r.sum(axis = 0) * M)   
```
